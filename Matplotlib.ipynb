```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y,'r*')
plt.xlabel('angle')
plt.ylabel('sine value')
```

#bar plot

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Hindi']
people=[100,50,150,25]
ax.bar(languages, people)
plt.xlabel('Language')
plt.ylabel('no of people')
plt.show()
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Spanish','Hindi']
people=[100,50,150,25]
ax.pie(people,labels=languages, autopct='%1.1f%%')
plt.show()

```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

#to plot 3d figure

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
